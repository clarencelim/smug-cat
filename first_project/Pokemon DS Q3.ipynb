### Q3 Which Pokemon can do the BIGGEST Damage?

The issue with this question is that each pokemon has 2 types. And their moves are roled on different stats depending on what type they are.

Eg. Fire -> Roles off Sp. Atk against Sp. Def (Special Type Moves)

Fighting -> Roles off Attack against Defense (Physical Type Moves)

Since each pokemon has 2 potential types we have 3 permutations. They can either be **pure physical (pp)**, **pure special (sp)** or **hybrid** (h).

A large component of this analysis was figuring out ways of how to slice data into these categories.

Furthermore, each pokemon has type advantage/weakness which multiplies it's stats against a said type. For sake of simplicity I assumed type advantage would multiply offensive stats by a factor of x2.

```python
import numpy as np
import pandas as pd
```

### Slicing of Data into 3 Subsets

```python
#dfp =  physical primary typset
dfp = df[df['Type 1'].str.contains('Normal|Fighting|Flying|Ground|Rock|Bug|Ghost|Poison|Steel')]
dfpp = dfp[dfp['Type 2'].str.contains('Normal|Fighting|Flying|Ground|Rock|Bug|Ghost|Poison|Steel', na = False)]
```

```python
#Purely Physical Pokemon
dfpp.head()
```

```python
#Physical Hybrid Pokemon
dfph = dfp[~dfp['Type 2'].str.contains('Normal|Fighting|Flying|Ground|Rock|Bug|Ghost|Poison|Steel', na = False)]
```

```python
dfph.head()
```

```python
#dfs = special primary typeset
dfs = df[~df['Type 1'].str.contains('Normal|Fighting|Flying|Ground|Rock|Bug|Ghost|Poison|Steel')]
dfs.head()
```

```python
dfps = dfs[~dfs['Type 2'].str.contains('Normal|Fighting|Flying|Ground|Rock|Bug|Ghost|Poison|Steel', na = False)]
dfps.head()
```

```python
dfsh = dfs[dfs['Type 2'].str.contains('Normal|Fighting|Flying|Ground|Rock|Bug|Ghost|Poison|Steel', na = False)]
dfsh.head()
```

```python
#checking if numbers tally
dfsh.count() + dfps.count() + dfpp.count() + dfph.count()
```

```python
#combining hybrid DFs
dfh = pd.concat([dfph ,dfsh], sort = False)
dfh.head()
```
