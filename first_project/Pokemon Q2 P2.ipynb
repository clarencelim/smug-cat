```python
filter = (df['Type 1'] == 'Water') | (df['Type 1'] == 'Fire')
```

```python
df[filter].reset_index()
```

```python
filter = ((df['Type 1'] == 'Water') | (df['Type 1'] == 'Fire')) & (df['Legendary'] == False)
```

```python
# strongest Water non legendary Menaphy! (total = 600), strongest Fire, Arcanine! (total = 555)
df[filter].sort_values('Total', ascending = False)
```

```python
#Can the weakest Water pokeman beat the strongest Fire pokeman even with elemental advantage?
df[filter].sort_values('Total', ascending = True)
```
