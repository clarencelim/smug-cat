```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('by_type.csv')
df
```

```python

plt.figure(figsize=(15,10))

plt.title('Average Stats By Type')
plt.xlabel('Type 1')
plt.ylabel('Stats')

plt.plot(df['Type 1'], df['Total'], '.-')

plt.legend()
plt.savefig('type_by_totalstats.png', dpi = 300)
plt.show()
```

```python
plt.figure(figsize=(15,10))


plt.plot(df['Type 1'], df['Attack'], '.')
plt.plot(df['Type 1'], df['HP'], '.')
plt.plot(df['Type 1'], df['Defense'], '.')
plt.plot(df['Type 1'], df['Speed'], '.')
plt.plot(df['Type 1'], df['Sp. Atk'], '.')
plt.plot(df['Type 1'], df['Sp. Def'], '.')
plt.legend()

plt.savefig('type_statbreakdown.png', dpi = 300)

plt.show()
```
