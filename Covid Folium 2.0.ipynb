# Bubble Map 2.0 Confirmed Cases/Deaths/Recoveries Layered

```python
import folium
from folium import plugins
import ipywidgets
import numpy as np
import pandas as pd
```

## Setting up Data

```python
df = pd.read_csv('datasets/covid_19_clean_complete_6_7_20.csv', parse_dates = ['Date'])
```

```python
df.head()
```

```python
latest = df['Date'] == max(df['Date'])
```

```python
df = df[latest]
```

```python
unmodified = df['Province/State'].isnull()
```

```python
df1 = df[unmodified]
```

```python
df1.set_index(['Country/Region'], inplace = True)
```

```python
df1.head()
```

```python
df2 = df.dropna()
```

```python
df2['Country/Region'].unique()
```

```python
df2.head()
```

```python
df3 = df2.groupby(['Country/Region']).sum()
```

```python
geo = pd.read_csv('worldgeodata.csv')
```

```python
geo.head()
```

```python
geo.drop('country',axis =1, inplace = True)
```

```python
geo.rename(columns= {'name' : 'Country/Region', 'latitude' : 'Lat', 'longitude':'Long'},inplace = True)
```

```python
geo.set_index(['Country/Region'], inplace = True)
```

```python
geo.head()
```

```python
geo.to_csv('lat_lon_by_country.csv', index = False)
```

```python
df3
```

```python
df3.index.tolist()
```

```python
geo.loc['China']['Lat']
```

```python
countries = df3.index.tolist()
```

```python
Lat = []
Long = []

for country in countries:
    Lat.append(geo.loc[country]['Lat'])

for country in countries:
    Long.append(geo.loc[country]['Long'])

print(Lat)
print(Long)
```

```python
df3['Lat'] = Lat
```

```python
df3['Long'] = Long
```

```python
df3
```

## Formatting the Map

```python
# map
map=folium.Map(location=[40, 40], zoom_start=2)

# plugin for mini map
minimap = plugins.MiniMap(toggle_display=True)

# add minimap to map
map.add_child(minimap)

# add scroll zoom toggler to map
plugins.ScrollZoomToggler().add_to(map)

# add full screen button to map
plugins.Fullscreen(position='topright').add_to(map)

#Confirmed Cases Layer

fg1 = folium.FeatureGroup(name = 'Confirmed Cases')

for lat,lon,country,confirmed in zip(df1['Lat'], df1['Long'], df1.index.tolist(), df1['Confirmed']):
    fg1.add_child(folium.Circle(popup=confirmed, 
                                tooltip = country,
                                radius= confirmed, 
                                location= (lat,lon),
                                fill = True,
                                fill_color = 'lightblue'))

for lat,lon,country,confirmed in zip(df3['Lat'], df3['Long'], df3.index.tolist(), df3['Confirmed']):
    fg1.add_child(folium.Circle(popup=confirmed, 
                                tooltip = country,
                                radius= confirmed, 
                                location= (lat,lon),
                                fill = True,
                                fill_color = 'lightblue'))

#Number of Deaths layer

fg2 = folium.FeatureGroup(name = 'Number of Deaths')
    
for lat,lon,country,deaths in zip(df1['Lat'], df1['Long'], df1.index.tolist(), df1['Deaths']):
    fg2.add_child(folium.Circle(popup=deaths, 
                                tooltip = country,
                                radius= deaths, 
                                location= (lat,lon),
                                color = 'red',
                                fill = True,
                                fill_color = 'red'))

for lat,lon,country,deaths in zip(df3['Lat'], df3['Long'], df3.index.tolist(), df3['Deaths']):
    fg2.add_child(folium.Circle(popup=deaths, 
                                tooltip = country,
                                radius= confirmed, 
                                location= (lat,lon),
                                color = 'red',
                                fill = True,
                                fill_color = 'red'))

#Number of recoveries

fg3 = folium.FeatureGroup(name = 'Number of Recoveries')
    
for lat,lon,country,recovered in zip(df1['Lat'], df1['Long'], df1.index.tolist(), df1['Recovered']):
    fg3.add_child(folium.Circle(popup=recovered, 
                                tooltip = country,
                                radius= recovered, 
                                location= (lat,lon),
                                color = 'green',
                                fill = True,
                                fill_color = 'lightgreen'))

for lat,lon,country,recovered in zip(df3['Lat'], df3['Long'], df3.index.tolist(), df3['Recovered']):
    fg3.add_child(folium.Circle(popup=recovered, 
                                tooltip = country,
                                radius= recovered, 
                                location= (lat,lon),
                                color = 'green',
                                fill = True,
                                fill_color = 'lightgreen'))
    
map.add_child(fg1)
map.add_child(fg2)
map.add_child(fg3)

map.add_child(folium.LayerControl())
```

### Notes

Here we have completed an improved version of the map with three layers showing no. of Cases, Deaths and Recoveries.

Hover over to find out which country it it, click to find out the respective number. Later control is in the top right.
